```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import openpyxl as opp
```

```python
vibration_data = pd.read_csv('dataset_02.csv', sep=';')
```

```python
tempo = vibration_data['t']
tensao = vibration_data['tensao']
vibracao = vibration_data['vibracao']
vibracaoabs = abs(vibration_data['vibracao'])
valor_inicial = 0
delta_t = 3.906250e-5
vibration_data['tempo'] = valor_inicial + (vibration_data.index * delta_t)
vibration_data.to_csv('vibration_datawtime', index = False)
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

plt.plot(vibration_data['tempo'],vibracaoabs)

plt.xlabel('Tempo(s)', fontsize = 19)
plt.ylabel('Vibração(g)', fontsize = 19)

# Adicionar um título ao gráfico
plt.title('Vibração X Tempo', fontsize = 19)

# Exibir o gráfico
plt.show()
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

plt.plot(vibration_data['tempo'],vibration_data['tensao'] )

plt.xlabel('tempo')
plt.ylabel('tensão')

# Adicionar um título ao gráfico
plt.title('Gráfico Tensão X Tempo')

# Exibir o gráfico
plt.show()
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

plt.plot(vibration_data['tempo'],vibration_data['vibracao'] )

plt.xlabel('tempo')
plt.ylabel('vibração')

# Adicionar um título ao gráfico
plt.title('Gráfico Vibração X Tempo')

# Exibir o gráfico
plt.show()
```

```python

# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

# Suponha que você tenha dois arrays de dados
dados1 = vibration_data['vibracao']
dados2 = vibration_data['tensao']

# Crie um array para o eixo x (opcional, neste caso usaremos índices)
eixo_x = vibration_data['tempo']

# Plote os dados 1 no eixo y
plt.plot(eixo_x, dados1, label='Vibração', )

# Plote os dados 2 no eixo y
plt.plot(eixo_x, dados2, label='Tensão')

# Adicione rótulos aos eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')



# Adicione uma legenda
plt.legend()

# Exiba o gráfico
plt.show()
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

# Suponha que você tenha dois arrays de dados
dados1 = vibration_data['tensao']
dados2 = vibracaoabs

# Crie um array para o eixo x (opcional, neste caso usaremos índices)
eixo_x = vibration_data['tempo']

# Plote os dados 1 no eixo y
plt.plot(eixo_x, dados1, label='Tensão',, fontsize = 30)

# Plote os dados 2 no eixo y
plt.plot(eixo_x, dados2, label='Vibração')

# Adicione rótulos aos eixos
plt.xlabel('Tempo (s)', fontsize = 30)
plt.ylabel('Tensão (V); Vibração (g)', fontsize = 30)
plt.xticks(fontsize=30)  # Aumenta o tamanho da fonte nos números do eixo x
plt.yticks(fontsize=30)  # Aumenta o tamanho da fonte nos números do eixo y
plt.gca().set_ylim(0, 5)  # Escala




# Adicione uma legenda
plt.legend()

# Exiba o gráfico
plt.show()
```
