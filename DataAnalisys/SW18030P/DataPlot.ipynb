```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import openpyxl as opp
```

```python
vibration_data = pd.read_csv('dataset_00.csv', sep=';')
```

```python
tempo = vibration_data['t']
tensao = vibration_data['tensao']
vibracao = vibration_data['vibracao']
valor_inicial = 0
delta_t = 3.906250e-5
vibration_data['tempo'] = valor_inicial + (vibration_data.index * delta_t)
vibration_data.to_csv('vibration_datawtime', index = False)
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

plt.plot(vibration_data['tempo'],vibration_data['tensao'] )

plt.xlabel('tempo')
plt.ylabel('tenso')

# Adicionar um ttulo ao grfico
plt.title('Grfico Tenso X Tempo')

# Exibir o grfico
plt.show()
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

plt.plot(vibration_data['tempo'],vibration_data['vibracao'] )

plt.xlabel('tempo')
plt.ylabel('vibrao')

# Adicionar um ttulo ao grfico
plt.title('Grfico Vibrao X Tempo')

# Exibir o grfico
plt.show()
```

```python

# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

# Suponha que voc tenha dois arrays de dados
dados1 = vibration_data['tensao']
dados2 = vibration_data['vibracao'] 

# Crie um array para o eixo x (opcional, neste caso usaremos ndices)
eixo_x = vibration_data['tempo']

# Plote os dados 1 no eixo y
plt.plot(eixo_x, dados1, label='Vibrao', )

# Plote os dados 2 no eixo y
plt.plot(eixo_x, dados2, label='Tenso')

# Adicione rtulos aos eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')



# Adicione uma legenda
plt.legend()

# Exiba o grfico
plt.show()
```
